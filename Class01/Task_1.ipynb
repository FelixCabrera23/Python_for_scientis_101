```python

def MC_e (N=1000):
    
    bins = np.zeros(N)
    
    for i in range(N):
        case = int(rand()*(N-1))
        bins[case]+=1
    
    non = np.count_nonzero(bins)
    
    Ee = N/(N-non)
    return(Ee)

MC_e(2)
```

```python
e = []
x = []
i = 3
while i < 1000:
    x.append(i)
    e.append(MC_e(i))
    i += 1

plt.plot(x,e)
plt.axhline(y=np.e, linestyle='--',color = 'orange')
```
