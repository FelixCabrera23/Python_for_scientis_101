## The area under the curve (a revision)

Consider the error function.
$$\text{erf}(z) = \frac{2}{\sqrt{\pi}} \int_{0}^{z} e^{-t^2} dt$$

This is known not to have an analytic solution. It has many uses in the physics fields and mathematics.

For us will be a great example to solve it numerically.

Consider erf(1.25)

```python
# importing usefull libraries
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#lets plot the integrand
x = np.linspace(0,1.25,100)
y = (2/np.sqrt(np.pi))*np.exp(-x**2)

plt.plot(x,y)
plt.show()
```

```python
# Solve it by Hit and miss
Z = 1.25
y_max = 2/np.sqrt(np.pi)
# N number of points
N = 1000
# generating random values
x_r = rand(N)*Z
y_r = rand(N)*y_max

# This logic finds the hits and store them
hits = y_r <= y_max*np.exp(-x_r**2)

# Fraction of hits
f = np.sum(hits)/N
area = f*(Z*y_max)
print(area)
```

```python
# lets do some fun!
plt.plot(x,y)
plt.scatter(x_r,y_r)
plt.show()
```
