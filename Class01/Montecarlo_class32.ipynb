## The area under the curve (a revision)

Consider the error function.
$$\text{erf}(z) = \frac{2}{\sqrt{\pi}} \int_{0}^{z} e^{-t^2} dt$$

This is known not to have an analytic solution. It has many uses in the physics fields and mathematics.

For us will be a great example to solve it numerically.

Consider erf(1.25)

```python
# importing usefull libraries
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#lets plot the integrand
x = np.linspace(0,1.25,100)
y = (2/np.sqrt(np.pi))*np.exp(-x**2)

plt.plot(x,y)
plt.show()
```

### Hit and miss approach

The idea is to generate random numbers in the known area and then verify if they are under the curve, the proportion of this points against the total points is the proportion of the area under the curve compared to the known total area. we can see that in a plot.

```python
# Solve it by Hit and miss
Z = 1.25
y_max = 2/np.sqrt(np.pi)
# N number of points
N = 1000
# generating random values
x_r = rand(N)*Z
y_r = rand(N)*y_max

# This logic finds the hits and store them
hits = y_r <= y_max*np.exp(-x_r**2)

# Fraction of hits
f = np.sum(hits)/N
area = f*(Z*y_max)
print(area)
```

```python
# lets do some fun!
x_in = np.select([y_r<=y_max*np.exp(-x_r**2)],[x_r])
y_in = np.select([y_r<=y_max*np.exp(-x_r**2)],[y_r])
plt.plot(x,y)
plt.scatter(x_r,y_r)
plt.scatter(x_in,y_in)

plt.show()
```

### Sample mean method

In this method, the area under a curve is even by the average value of the curve multiplied by the range over which it is integrated.

We have to calculate the value of the curve multiplied by the range over a number of values $x$ taken randomly between 0 and 1.25, then take the arithmetic average of these values and multiply the result by the range of integration. This is the 'Sample Mean' Monte Carlo estimate of the integral.


```python
Z = 1.25
y_max = 2/np.sqrt(np.pi)

N = 500

x_m = rand(N)*Z
y_m = y_max*np.exp(-x_m**2)

y_ave = np.mean(y_m)

area = y_ave * Z

print(area)
```

```python
# lets visualise
plt.plot(x,y)
plt.scatter(x_m,y_m)
plt.show()
```

```python
# Lets make this 2 into functions so we can extract some more data!
    
Z = 1.25
erf_z = lambda z : (2/np.sqrt(np.pi))*np.exp(-z**2)

# Solve it by Hit and miss
def hit_miss (func,x_0,x_f,N):

    x = np.linspace(x_0,x_f,100)
    y = func(x)
    y_max = np.max(y)
    x_r = rand(N)*(x_f-x_0)+x_0
    y_r = rand(N)*y_max

    hits = y_r <= func(x_r)

    f = np.sum(hits)/N
    area = f*(abs(x_f-x_0)*y_max)

    return float(area)

hit_miss(erf_z,0,Z,1000)
```
