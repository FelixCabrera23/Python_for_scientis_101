# Optimization

We have seen how the idea of using random numbers is very useful to find the numerical solution of integrals. We can also try to use this idea to solve the inverse problem and use it to optimize a function, meaning to finding a minimum solution in a delimited area. There are many different forms to implement this method to solve this problem, we will try to understand two of them. Random walks and semi-random sampling.

## Semi-random sampling

Let's consider the case of a quadratic equation. The solution is unique and trivial for the use of python to find it, however will work to consider if our method works.

Consider the equation:
$$ y = x^2 $$

```python
# importing usefull libraries
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Defining quadratic function
def Quad (x, a = 1, b = 0, c = 0):
    return a*x**2 + b*x + c
```

```python
x = np.linspace(-20,20,100)
y = Quad(x)

plt.plot(x,y)
```

With our power of generating random numbers, we can generate a set and then test for the minimum result. Let's consider the range from -20 to 20 in the $x$ axis.

```python
# Consider the range -20 to 20
N = 100
x_rand = rand(N)*40 - 20
y_rand = Quad(x_rand)

plt.plot(x,y)
plt.scatter(x_rand,y_rand)
plt.show()
```

In this case we are using numpy functions instead of integrated Python functions to search for the minimum since we are using numpy arrays and if we were to use Python min function it will be less efficient.

```python
y_min = np.min(y_rand)
x_min = x_rand[np.where(y_rand == y_min)[0][0]]
print (y_min)
print (x_min)
```

We can observe that just randomly sampling we will need to perform a considerable amount of sampling, we can improve this if after a certain amount of samples we reduce the area of sampling.
Let's consider the idea of reducing the range by steps. In this form we can use fewer samples but in a more efficient range. 

The idea is the as follows: 
- Generate random values for $x$, let's say $N=100$, in the range $(L1,L2)$.
- Determine which is the minimum result for this set.
- Change the center and compute the new limits of the set $(L1,L2)$
- Repeat the process an $R$ number of times (Runs)

```python
N = 100
C = -10
for i in range(N//10):
    L = 2*((20)-i*(20/10))
    x_rand = rand(N//10)*L-(L/2)+C
    y_rand = Quad(x_rand)
    y_min = np.min(y_rand)
    C = x_rand[np.where(y_rand == y_min)[0][0]]
    plt.scatter(x_rand,y_rand,label= i+1) # we print as legend the run number
    print(L,C)

plt.legend()
plt.show()
```
